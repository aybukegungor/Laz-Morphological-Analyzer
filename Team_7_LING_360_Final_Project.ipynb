```python
def character_changer(input_word, lower=True):
  """Takes in a string, if any character in the string is in the Laz alphabet but not in the Turkish alphabet, replaces it with the Turkish alternative."""
  characters = {"č": "ç'", "ǯ": "ts", "ʒ": "dz", "çh": "ç'", "kh": "k'", "th": "t'", "ph": "p'", "zh": "dz", "ṕ": "p'", "ǩ": "k'", "t̆": "t'", "ť": "t'", "z̆": "z'"}
  for character in input_word:
    if character in characters:
      input_word = input_word.replace(character, characters[character])
  if(lower):
    input_word = input_word.lower()
  return input_word
```

```python
def analyzer():
  """Takes in an input, if any character in the input is in the Laz alphabet but not in the Turkish alphabet, replaces it with the Turkish alternative using the character_changer() function.
     Then returns the morphological analyses on the optimized input as output, using the laz_analyzer.lookup() function.
     If the input is "q", breaks."""
  while True:
        input_word = input("Xela k'aoba! Please enter the word you want to analyze or press 'q' to exit.")
        if input_word == "q":
            return "Guriş xelate!"
        else:
            output = character_changer(input_word)
            output = laz_analyzer.lookup(output)
        print(output)
```

```python
def generator():
  """Takes in an input, if any character in the input is in the Laz alphabet but not in the Turkish alphabet, replaces it with the Turkish alternative using the character_changer() function.
     Then returns the generated version of the optimized input as output, using the laz_generator.lookup() function.
     If the input is "q", breaks."""
  while True:
        input_word = input("Xela k'aoba! Please enter the word you want to generate or press 'q' to exit.")
        if input_word == "q":
            return "Guriş xelate!"
        else:
            output = character_changer(input_word, False)
            output = laz_generator.lookup(output)
        print(output)
```
